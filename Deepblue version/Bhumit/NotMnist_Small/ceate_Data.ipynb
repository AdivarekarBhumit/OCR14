```python
# Create image data
path = os.getcwd()
train_dir = path + '/Data'
train_dir_list = os.listdir(train_dir)
```

```python
train_dir_list
```

```python
train_images = []
label = 0 
labels = []

for folder in train_dir_list:
    img_list = os.listdir(train_dir + '/' + folder)
    print('Working in folder ', folder)
    for img in img_list:
        ip_img = cv2.imread(train_dir + '/' + folder + '/' + img, cv2.IMREAD_GRAYSCALE)
        labels.append(label)
        print(ip_img.shape)
        train_images.append(ip_img)
    label += 1
```

```python
train_data = np.array(train_images)
```

```python
train_data.shape
```

```python
train_data = train_data.astype('float32')
```

```python
#Normalize
train_data /= 255
```

```python
train_data.shape
```

```python
train_data = train_data.reshape(18724,28,28,1)
```

```python
train_data.shape
```

```python
labels = np.array(labels)
```

```python
#save array files
np.save('train_data.npy', train_data)
np.save('labels.npy', labels)
```
